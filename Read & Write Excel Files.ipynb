#### Reading And Writing Xlsx files in pyspark
*  We can use `com.crealytics.spark.excel`  library for reading and writing xls files in pyspark
* while reading or writing xls files we need to use header option `pation("header","true")`

```
pip install spark-excel_2.12
```

```
excel_data=spark.read.format("com.crealytics.spark.excel").option("header", "true").option("inferSchema", "true").load("/FileStore/tables/emp.xlsx")
```

```
excel_data.printSchema()
```

```
display(excel_data)
```

```
emp_csv=spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/FileStore/tables/emp.csv")
```

```
display(emp_csv)
```

#### Writing data into xlsx file use spark library  `com.crealytics.spark.excel`

```
emp_csv.select("*").write.format('com.crealytics.spark.excel').option("header","true").option("inferSchema", "true").save('/FileStore/tables/excelfiles',mode="overwrite")
```
