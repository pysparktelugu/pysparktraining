#### Spark Session Creation
 * A SparkSession can be used create DataFrame, register DataFrame as tables, execute SQL over tables, cache tables, and read parquet files. 
 * To create a SparkSession, use the following builder pattern:

```
from pyspark.sql import SparkSession 
#from pyspark.conf import SparkConf

#spark1 = SparkSession.builder.master("local").appName("databricks spark").getOrCreate()
no_cpu=2
spark = SparkSession.builder.appName('localSpark')\
.master('local[{}]'.format(no_cpu))\
.config("spark.driver.memory", "12g")\
.config('spark.executor.cores', '3')\
.config('spark.cores.max', '3')\
.getOrCreate()
```

##### SET Configuration parameters in spark

```
conf = spark.sparkContext._conf.setAll([('spark.executor.memory', '2g'), 
                                        ('spark.app.name', 'Spark Updated Conf'), 
                                        ('spark.executor.cores', '4'), 
                                        ('spark.cores.max', '4'), 
                                        ('spark.driver.memory','4g')])
```
